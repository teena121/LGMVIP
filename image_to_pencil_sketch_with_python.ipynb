```python
from google.colab.patches import cv2_imshow
```

**Import the python library**

```python
#import the required library
import cv2
```

```python
#read the image in RGB color format which we want to covert into pencil sketch from the correct path
img = cv2.imread("/monalisa.jpg")
```

**Finding and displaying the image we want to convert into pencil sketch**

```python
#Here we display the real image
cv2_imshow(img)
```

**Now we convert the image into gray scale image means a classic black and white picture!!**

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#display the gray scale image
cv2_imshow(img_gray)
```

**Now here we invert the obtained gray scale image which we call as inverted gray scale image**

```python
img_invert = cv2.bitwise_not(img_gray)
```

```python
#display the inverted gray scale image
cv2_imshow(img_invert)
```

**Now we use the GaussianBlur to get more blur image**

```python
img_softblur = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(img_softblur)
```

**Here we get the blur picture of the inverted image**

```python
img_blur_inverted = 255-img_softblur
cv2_imshow(img_blur_inverted)
```

**Here comes the final step to create a pencil sketch by dividing gray scale image by inverted blur image**

```python
pencil_sketch_img = cv2.divide(img_gray,img_blur_inverted,scale=250)
cv2_imshow(pencil_sketch_img)
```

**And the desired output is obtained.**
